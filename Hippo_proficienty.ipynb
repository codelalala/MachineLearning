```python
clean_data=df.copy()
def s(data):
    if(data.find('GE'))==0:
        data=data[2:4]
        #print(data)
    if(data.find('LT'))==0:
        data=str(int(data[2:4])-1)
    if(data.find('LE'))==0:
        data=data[2:4]
    if(data.find('GT'))==0:
        data=str(int(data[2:4])+1)
    if(data.find('-'))==2:
        #print(data[0:2],data[3:5],int((int(data[0:2])+int(data[3:5]))/2))
        data=str(int((int(data[0:2])+int(data[3:5]))/2))
    return data
                              
clean_data.ALL_MTH00PCTPROF_1415=clean_data.ALL_MTH00PCTPROF_1415.apply(s)

clean_data.tail(10)
```
